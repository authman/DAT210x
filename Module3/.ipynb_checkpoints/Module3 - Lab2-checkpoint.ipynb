```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Look pretty...

# matplotlib.style.use('ggplot')
plt.style.use('ggplot')
```

Load up the wheat seeds dataset into a dataframe. We've stored a copy in the Datasets directory.

```python
df = pd.read_csv('Datasets\\wheat.data')
```

Create a 2d scatter plot that graphs the `area` and `perimeter` features:

```python
df2 = df[['area','perimeter']]
df2
df2.plot.scatter(x='area',y='perimeter',marker='x')
plt.show()
```

Create a 2d scatter plot that graphs the `groove` and `asymmetry` features:

```python
df3 = df[['groove','asymmetry']]
df3
df3.plot.scatter(x='groove',y='asymmetry',marker='+')
plt.show()
```

Create a 2d scatter plot that graphs the `compactness` and `width` features:

```python
df4 = df[['compactness','width']]
df4
df4.plot.scatter(x='compactness',y='width', marker='*')
plt.show()
```
